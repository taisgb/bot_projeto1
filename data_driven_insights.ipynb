##Execução

```python
#Listas, Dicionários e Tuplas:

# Lista: Extraindo a coluna de idade dos alunos
idades = list(df['age'])
print("As 5 primeiras idades da lista:")
for idade in idades[:5]:
    print(idade)

# Dicionário: Relacionando idade e sexo dos 5 primeiros alunos
# Usamos .iloc para garantir que pegamos os 5 primeiros registros
dicionario_aluno_sexo = dict(zip(df['age'].iloc[:5], df['sex'].iloc[:5]))
print("\nDicionário relacionando idade e sexo:")
print(dicionario_aluno_sexo)

# Tupla: Contendo três informações distintas da primeira linha (idade, sexo, escola)
primeira_linha = df.iloc[0]
tupla_aluno = (primeira_linha['age'], primeira_linha['sex'], primeira_linha['school'])
print("\nTupla com informações do primeiro aluno:")
print(tupla_aluno)
```

```python
# Estruturas Condicionais e Laços

# Coluna numérica de interesse: Nota final (G3)
df['G3'] = pd.to_numeric(df['G3'])
coluna_notas = 'G3'
primeira_nota = df[coluna_notas][0]

# Estrutura condicional para avaliar a primeira nota
print(f"A primeira nota da lista é: {primeira_nota}")
if primeira_nota >= 10:
    print("Status: Aprovado(a)")
else:
    print("Status: Reprovado(a)")

# Laço 'for' para calcular a média das 5 primeiras notas
soma_notas = 0
lista_primeiras_notas = list(df[coluna_notas][:5])
for nota in lista_primeiras_notas:
    soma_notas += nota
media = soma_notas / len(lista_primeiras_notas)
print(f"\nA média das 5 primeiras notas é: {media}")

# Laço 'while' para encontrar a primeira nota perfeita (20)
i = 0
encontrou_nota_max = False
notas = list(df[coluna_notas])
while i < len(notas):
    if notas[i] == 20:
        print(f"\nEncontrada nota 20 na posição {i} da lista!")
        encontrou_nota_max = True
        break
    i += 1

if not encontrou_nota_max:
    print("\nNenhum aluno com nota 20 encontrado na base.")
```

```python
#Operadores Aritméticos e Manipulação de Dados

# Criando uma coluna com a média das notas (G1, G2, G3)
df['G_AVG'] = (df['G1'] + df['G2'] + df['G3']) / 3

# Criando uma coluna de status final ('Approved' ou 'Failed')
# Usaremos uma função e o método .apply()
def get_status(nota):
    if nota >= 10:
        return 'Approved'
    else:
        return 'Failed'

df['Final_Status'] = df['G3'].apply(get_status)

# Exibindo as novas colunas junto com as notas
print(df[['school', 'age', 'G1', 'G2', 'G3', 'G_AVG', 'Final_Status']].head())
```

```python
#NumPy e Arrays Numéricos

#Coluna de faltas ('absences')
array_faltas = np.array(df['absences'])

total_aulas = 93
percentual_faltas = (array_faltas / total_aulas) * 100

# Acessando e exibindo partes do array (fatiamento)
print("Array de faltas (10 primeiros alunos):", array_faltas[:10])

# Operação agregada sobre o array
print("\nSoma total de faltas de todos os alunos:", array_faltas.sum())
print("Média de faltas por aluno:", round(array_faltas.mean(), 2))
print("Máximo de faltas de um único aluno:", array_faltas.max())
```

```python
#Acesso e Manipulação de Dados com Pandas

# Selecionando alunos que pretendem cursar ensino superior ('higher' == 'yes')
alunos_ensino_superior = df.loc[df['higher'] == 'yes']
print(f"Número de alunos que querem ensino superior: {len(alunos_ensino_superior)}")
print("Média da nota final (G3) deste grupo:", round(alunos_ensino_superior['G3'].mean(), 2))

# Contando a distribuição de profissões das mães
print("\nDistribuição de profissões das mães (Mjob):")
print(df['Mjob'].value_counts())

# Análise agregada: Média da nota final (G3) por tempo de estudo (studytime)
media_por_tempo_estudo = df.groupby('studytime')['G3'].mean()
print("\nMédia da nota final por tempo de estudo semanal:")
print(media_por_tempo_estudo)
# Legenda: 1: <2 horas, 2: 2 a 5 horas, 3: 5 a 10 horas, 4: >10 horas
```

```python
# Gráfico de Linha para mostrar a distribuição das notas finais (G3)
contagem_notas = df['G3'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
contagem_notas.plot(kind='line', marker='o')
plt.title('Distribuição de Notas Finais (G3)')
plt.xlabel('Nota Final')
plt.ylabel('Número de Alunos')
plt.grid(True)
plt.xticks(range(0, 21, 2))
plt.show()
```

```python
# Gráfico de Barras (Média de G3 por educação da mãe)
media_por_edu_mae = df.groupby('Medu')['G3'].mean()

plt.figure(figsize=(10, 6))
media_por_edu_mae.plot(kind='bar', color='cornflowerblue')
plt.title('Média da Nota Final (G3) vs. Educação da Mãe')
plt.xlabel('Nível Educacional da Mãe (0=Nenhum, 4=Superior)')
plt.ylabel('Média da Nota Final')
plt.xticks(rotation=0)
plt.show()
```

```python
# Gráfico de Dispersão (relação entre faltas e nota final)
plt.figure(figsize=(10, 6))
plt.scatter(df['absences'], df['G3'], alpha=0.5) # alpha para ver sobreposição
plt.title('Dispersão: Faltas vs. Nota Final (G3)')
plt.xlabel('Número de Faltas')
plt.ylabel('Nota Final')
plt.grid(True)
plt.show()
```

# Relatório Final: O que os dados nos contam sobre a jornada de um estudante?

Neste projeto, analisei um conjunto de dados com informações de 649 jovens em Portugal, olhando para tudo: desde suas notas e hábitos de estudo até o ambiente familiar. Meu objetivo era encontrar padrões, ou seja, pistas que nos ajudassem a entender o que contribui para o sucesso na escola.

### Principais Descobertas:

1.  **O Tempo de Estudo é Crucial:** A análise mostrou uma correlação direta e forte entre o tempo semanal dedicado aos estudos (`studytime`) e a nota final (`G3`). Alunos que estudam mais, de fato, tiram notas maiores.

2.  **Influência Familiar:** O nível educacional dos pais, especialmente da mãe (`Medu`), tem um impacto positivo visível nas notas dos alunos. Isso sugere que o ambiente familiar e o incentivo aos estudos em casa são determinantes para o desempenho escolar.

3.  **Assiduidade:** Embora a maioria dos alunos tenha poucas faltas, aqueles com um número excessivo de ausências (`absences`) estão concentrados no grupo de menor desempenho. Manter a frequência nas aulas é um indicador chave para evitar a reprovação.

4.  **Aspirações Futuras:** A grande maioria dos alunos (90%) aspira ingressar no ensino superior, e esse grupo tende a ter um desempenho ligeiramente melhor. Essa meta futura parece funcionar como um fator motivacional.

Este projeto demonstrou como a análise de dados pode transformar números brutos em insights valiosos sobre educação, apontando para áreas onde intervenções (como programas de reforço de estudo ou de combate ao absenteísmo) poderiam ser mais eficazes.
